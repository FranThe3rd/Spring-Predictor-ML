```python
import pandas as pd
```

### **ENSO Pivoter**

```python
enso_data = pd.read_csv('../ENSO.csv', parse_dates=['date'])
enso = enso_data.copy()
enso['year'] = enso.date.dt.year
enso['month'] = enso.date.dt.month
enso['month_str'] = enso.date.dt.strftime('%b')
enso = enso.drop(columns=['date'])
enso_pivoted = enso.pivot(index='year', columns='month')
enso_pivoted.columns = ["enso_" + "_".join(str(x) for x in a) for a in enso_pivoted.columns.to_flat_index()]
```

### **Soil Pivoter**

```python
# **Soil Pivoter**
soil_data = pd.read_csv('../soil.csv', parse_dates=['day'])
soil = soil_data.copy()
soil['year'] = soil.day.dt.year
soil['day_of_year'] = soil.day.dt.day_of_year
soil = soil.drop(columns=['day']).pivot(index='year', columns='day_of_year')
soil_pivoted = soil.copy()
soil_pivoted.columns = ["soil_" + "_".join(str(x) for x in a) for a in soil_pivoted.columns.to_flat_index()]
```

### **Weather Pivoter**

```python
# **Weather Pivoter**
weather_data = pd.read_csv('../weather.csv', parse_dates=['date'])
weather = weather_data.copy()
weather['year'] = weather.date.dt.year
weather['day_of_year'] = weather.date.dt.day_of_year
weather = weather.drop(columns=['date']).pivot(index='year', columns='day_of_year')
weather_pivoted = weather.copy()
weather_pivoted.columns = ["weather_" + "_".join(str(x) for x in a) for a in weather_pivoted.columns.to_flat_index()]
```

```python

combined_data = enso_pivoted.merge(soil_pivoted, on='year', how='outer').merge(weather_pivoted, on='year', how='outer')

combined_data.to_csv('combined_pivoted_data.csv')

combined_data.head()
```
